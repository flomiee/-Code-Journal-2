```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data=np.random.uniform(0,1,1000)
plt.hist(data,bins=100,edgecolor='black',color='skyblue')
plt.title('Histogram of Uniform Distribution (100 samples, 100 bins)')
plt.xlabel('Value (Uniformly Distributed)')
plt.ylabel('Frequency')
plt.savefig('uniform_histogram.pdf'
```
