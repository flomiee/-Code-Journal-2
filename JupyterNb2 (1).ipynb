```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)
#exp(x) is exponential data so def that
def calculate_exponential(data):
    return np.exp(data)
```

```python
y = calculate_exponential(x)
plt.figure(figsize=(8,5))
#plot the horizontal axis and vertical axis
plt.plot=(x,y)
#then define
plt.xlabel("Time[milliseconds]")
plt.ylabel("Awesomeness")
plt.savefig('awesome_time.pdf')
plt.show()
```
