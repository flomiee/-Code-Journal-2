```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,1,100)
def get_sin(data):
    return np.sin(data)
def get_cos(data):
    return np.cos(data)
y_sin = get_sin(x)
y_cos = get_cos(x)
```

```python
fig, (ax1,ax2) = plt.subplots(1,2, figsize=(10,4))
ax1.plot(x, y_sin)

ax1.set_title('sin(x)')
ax1.set_xlabel('x')
ax1.set_ylabel('sin(x)')
ax1.grid(True)

ax2.plot(x, y_cos)
ax2.set_title('cos(x)')
ax2.set_xlabel('x')
ax2.set_ylabel('cos(x)')
ax2.grid(True)

plt.tight_layout()
plt.savefig('trig_functions.pdf')
```
