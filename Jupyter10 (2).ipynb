```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.random.gamma(3,20,1000)
plt.hist(data,bins=100,edgecolor='black',color='skyblue')
plt.title('Gamma Distribution over Continuous Positive Integer Variables')
plt.xlabel('Continuous Positive Integer Variables')
plt.ylabel('Gamma Distribution')
plt.savefig('gamma_distribution.pdf', format ='pdf')
plt.show()
```
